```python
# 불쾌지수 계산 함수
def calculate_discomfort_index(temp, humidity):
    return 0.81 * temp + 0.01 * humidity * (0.99 * temp - 14.3) + 46.3

# 적용
df_all['불쾌지수'] = calculate_discomfort_index(df_all['평균기온(℃)'], df_all['상대습도'])
```

```python
def 불쾌지수_범주화(di):
    if di < 75:
        return '쾌적'
    elif di < 80:
        return '약간 불쾌'
    elif di < 85:
        return '불쾌'
    else:
        return '매우 불쾌'

# 컬럼 생성
df_all['불쾌지수_등급'] = df_all['불쾌지수'].apply(불쾌지수_범주화)
```

```python
def calc_apparent_temp(T, V):
    return 13.12 + 0.6215 * T - 11.37 * (V ** 0.16) + 0.3965 * T * (V ** 0.16)

df_all['체감온도'] = calc_apparent_temp(df_all['평균기온(℃)'], df_all['평균풍속(m/s)'])
```

```python
def 체감온도_범주화(diz):
    if diz < 27:
        return '쾌적'
    elif diz < 31.9:
        return '주의'
    elif diz < 40.9:
        return '경고'
    elif diz < 53.9:
        return '심각'
    else:
        return '매우 위험'

# 컬럼 생성
df_all['체감온도_등급'] = df_all['체감온도'].apply(체감온도_범주화)
```
